```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
from sklearn.datasets import make_regression

x, y = make_regression(n_samples=100, n_features=5, noise=2)

print(f"shape of the data {x.shape}")
print(f"shape of the target {y.shape}")
print(f"No of samples {x.shape[0]}")
print(f"No of features {x.shape[1]}")

print(
    f"Similar cardinality in number of samples and No of values in the target :{'Matched' if len(x) == len(y) else 'Not Matched'}"
)
```

```python
df=pd.DataFrame(x,y)
sns.pairplot(df)
```

```python
from sklearn.model_selection import train_test_split as tts_function
from sklearn.impute import SimpleImputer as njan_imputer
from sklearn.linear_model import LinearRegression as LogisticRegression

alpha,beta,gamma,delta=tts_function(x,y,train_size=0.8)
model=LogisticRegression(n_jobs=-1)
what_to_type_here,what_should_I_put_here=alpha,gamma
model.fit(what_to_type_here,what_should_I_put_here)
```
