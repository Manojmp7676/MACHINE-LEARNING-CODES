```python
def formula(x):
    return 5 * x + 15

x=[_ for _ in range(10)]
y=[formula(i) for i in x]

for _ in zip(x,y):
       print(f"x: {_[0]},y={_[-1]}")

    
    
```

```python
formula = lambda x: 2 * x + 10

x = list(range(10))
y = list(map(formula, x))

for a, b in zip(x, y):
    print(f"x: {a}, y={b}")
```
