```python
import numpy as np
```

```python
np.random.seed(861)  # for reproducibility
X = np.random.randint(1,100000,size=5)
print(X)
```

```python
from sklearn.preprocessing import StandardScaler
s=StandardScaler()
X_scaled=s.fit_transform(X.reshape(-1,1))
print(X_scaled)
```

```python
s,n=0,0
for i in X:
    s+=i
    n=n+1
mean=(s/n)
print(mean)
```

```python
sq=0
for i in X:
    sq=sq+(i-mean)**2
std=(sq/n)**0.5
print(std)
```

```python
X_scaled=[]
for i in X:
    X_scaled.append((i-mean)/std)
print(X_scaled)
```

```python
from sklearn.preprocessing import MinMaxScaler
m=MinMaxScaler()
X_scaled=m.fit_transform(X.reshape(-1,1))
print(X_scaled)
```

```python
min=np.min(X)
max=np.max(X)
min,max=X[0],X[0]
for i in X:
    if i<min:
        min=i
    if i>max:
        max=i
print(min,max)        
```

```python
X_scaled2=[]
for i in X:
    X_scaled2.append(((i-min)/(max-min)))
print(X_scaled2)
```
