```python
import numpy as np
import matplotlib.pyplot as plt

X=np.array([10,20,30,40,50])
Y=np.array([12,15,19,31,43])

X_mean=np.mean(X)
Y_mean=np.mean(Y)

numerator=np.sum((X-X_mean)*(Y-Y_mean))
denominator=np.sum((X-X_mean)**2)

w=numerator/denominator
b=Y_mean-w*X_mean

Y_pred=w*X+b

print("slope (w):", w)
print("intercept (b):", b)

plt.scatter(X,Y)
plt.plot(X,Y_pred,color='red')
plt.grid(True)
plt.show()
```
