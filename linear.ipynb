create a function my_slope(a,b) which would accept two point a,b as an input such as a=(xi,yi)and b=($xj,yj)where xi,yi,xj,yj denotes the value in x , y

```python
def my_slope(a, b):
    dy=a[1]-b[1]
    dx=a[0]-b[0]
    return  dy/dx
```
